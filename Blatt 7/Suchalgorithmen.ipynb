```python
import random
import time

import numpy as np
from numpy.random import randint

import pandas as pd

import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#lineare Suche

def linearSearch(e,n,a):
    for i in range(0,n):
        if(a[i] == e):
            return i
    return -1

#Komplexität: best case: O(1); average: O(n/2); worst case: O(n)
```

```python
#binäre Suche
#Vorbedingung: Array a ist sortiert: a[i] < a[i+1], für alle i in [0,n-2]

#divide and conquer algorithmus

def binarySearch(e,n,a):
    min = 0
    max = n - 1
    while(min <= max):
        i = int(min + (max - min) / 2)
        if(a[i] == e):
            return i
        elif(a[i] < e):
            min = i + 1
        else:
            max = i - 1
    return -1

#Komplexität: best case: O(1); average case: O(log2(n)) - 1; worst case: O(log2(n))
```

```python
#Interpolationssuche

def interpolationSearch(e,n,a):
    min = 0
    max = n - 1
    while(min <= max):
        if(a[min] == a[max] == e):
            return a[min]
        elif(a[min] == a[max] != e):
            return -1
        
        i = int(min + (max - min) * (e - a[min]) / (a[max] - a[min]))
        if(a[i] == e):
            return i
        elif(a[i] < e):
            min = i + 1
        else:
            max = i - 1
    return -1

#Komplexität: best case: O(1); average case: O(log2(log2(n))); worst case: O(n)
```

```python
count = 1
z = 1
runtime_linear = []*10
while(count <= 10):

    random.seed(time.time())
    z*=5
    
    a = np.sort(randint(0,z, size=z))            
    n = len(a)
    e = a[random.randint(0,len(a)-1)]
    
    #Lineare Suche
    start = time.time()
    ergebnis = linearSearch(e,n,a)
    end = time.time()
    runtime_linear.append(end-start)
    print ("Lineare Suche, #%i\nIntervall: [%i, %i]\nErgebnis:%i" % (count, 0, z, ergebnis))
    print ("Laufzeit: ", (end-start), "\n")
    print ("\n")
    
    count += 1

x = []
s = 5
for num in range(1,11):
    s *= 5
    x.append(s)
y = runtime_linear

plt.plot(x, y)
plt.ylabel('Laufzeit')
plt.xlim(-1000000, 50000000)
plt.xlabel('Anzahl Elemente im Array')
plt.title('Laufzeitkomplexität der linearen Suche')
plt.show()
```

```python
#############################################################################

#Zusätzlich, nicht teil der Vorlesung/Hausaufgabe

#Quadratische Binärsuche

from math import sqrt

def quadBinarySearch(e,n,a):
    min = 0
    max = n - 1
    while(min <= max):
        if(a[min] == a[max] == e):
            return a[min]
        elif(a[min] == a[max] != e):
            return -1
        
        i = int(min + (max - min) * (e - a[min]) / (a[max] - a[min]))
        if(a[i] == e):
            return i
        elif(a[i] > e):
            while(a[i] > e):
                i -= int(sqrt(n))
            min = int(i + sqrt(n) - 1)
        else:
            while(a[i] < e):
                i += int(sqrt(n))
            max = int(i - sqrt(n) + 1)
    return -1

#Komplexität: best case: O(1); average case: O(log2(log2(n))); worst case: O(sqrt(n))
```

```python
count = 1
z = 1
runtime_quad = []*10
while(count <= 10):
    
    random.seed(time.time())
    z*=5
    
    a = np.sort(randint(0,z, size=z))            
    n = len(a)
    e = a[random.randint(0,len(a)-1)]
    
    #Quadratische Binärsuche
    start = time.time()
    ergebnis = quadBinarySearch(e,n,a)
    end = time.time()
    runtime_quad.append(end-start)
    print ("quadBinarySearch, #%i\nIntervall: [%i, %i]\nErgebnis:%i" % (count, 0, z, ergebnis))
    print ("Laufzeit: ", (end-start), "\n")
    count += 1
    
x = []
s = 5
for num in range(1,11):
    s *= 5 
    x.append(s) 
y = runtime_quad


plt.plot(x, y)
plt.xlim(-1000000, 50000000)
plt.ylabel('Laufzeit')
plt.xlabel('Anzahl Elemente im Array')
plt.title('Laufzeitkomplexität der Quadratishen Binärsuche')
plt.show()
```
