```python
import time
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline

def primzahlen_liste(limit):
    a = [True] * limit                          
    a[0] = a[1] = False

    for (i, isprime) in enumerate(a):
        if isprime:
            yield i
            for n in range(i*i, limit, i):     
                a[n] = False
                


                
count = 1
limit = 100
runtime = []
test = []
while(count <= 10):
    
    limit*=100000
    
    start = time.time()
    test = primzahlen_liste(limit)
    end = time.time()
    runtime.append(end-start)
    
    
    count += 1
                
x = []
s = 100

for num in range(1,11):
    s *= 10000 
    x.append(s) 
y = runtime


plt.plot(x, y)
plt.xlim(-1000000, 50000000)
plt.ylabel('Laufzeit')
plt.xlabel('Limit')
plt.title('Laufzeitkomplexität')
plt.show()

#komplexität sieht nach log log n aus ?
#speicherkomplexität: O(n)
```
